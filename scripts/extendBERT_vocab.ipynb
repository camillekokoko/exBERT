```python
import spacy

# Load the finetuned spaCy language model
nlp1 = spacy.load(r"/Users/camilleko/NLP_summarization/NER/output/model-best")

org_name_path = '/Users/camilleko/NLP_summarization/data/interim/Organization_Name.txt'

with open(org_name_path, "r", encoding="utf-8") as file:
        lines = file.readlines()
        file_contents = ''.join(lines[1:])
        file_contents = file_contents.replace('\n', ' ')

print(len(file_contents))

# Process the document with spaCy
doc = nlp1(file_contents)

# Extract tokens
tokens = [token.text for token in doc]
print(len(tokens))

unique_tokens = list(set(tokens))

print(len(unique_tokens))
```

```python
from transformers import BertTokenizer

# Load the BERT tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# Access the vocabulary as a Python list
vocabulary_list = tokenizer.get_vocab()

print(len(vocabulary_list))
```

```python
# Initialize a list to store the results
results = []

# Check if each new token exists in the vocabulary
for token in lowercase_unique_tokens:
    if token in vocabulary_list:
        results.append(f"'{token}' exists in the vocabulary.")
    else:
        results.append(f"'{token}' does not exist in the vocabulary.")

# Print the results
for result in results:
    print(result)
print(len(results))
```

```python
from transformers import BertTokenizer, BertForMaskedLM

# Load the BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertForMaskedLM.from_pretrained("bert-base-uncased")

# Add new tokens to the tokenizer's vocabulary
tokenizer.add_tokens(lowercase_unique_tokens)

# You can also resize the model's embeddings matrix accordingly
model.resize_token_embeddings(len(tokenizer))



```

```python
print(tokenizer)
```

```python
# Access the existing vocabulary as a Python dictionary
existing_vocab = tokenizer.get_vocab()

# Combine the new tokens with the existing vocabulary
combined_vocab = existing_vocab.copy()

for token in lowercase_unique_tokens:
    if token not in existing_vocab:
        combined_vocab[token] = len(combined_vocab)

# Save the combined vocabulary to a text file
output_file_path = "custom-bert-vocab.txt"

with open(output_file_path, "w", encoding="utf-8") as output_file:
    for token, token_id in combined_vocab.items():
        output_file.write(f"{token}\n")

print(f"Custom vocabulary saved to '{output_file_path}'.")
```
